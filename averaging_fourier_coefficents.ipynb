```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

ntrials = 100
srate = 200
time = np.arange(0,1,1/srate)
pnts = len(time)

data = np.zeros((ntrials,pnts))

for i in range(ntrials):
     data[i,:] = np.sin(2*np.pi*20*time + 2*np.pi*np.random.rand())

#plot the data
for j in range(ntrials):
     plt.plot(time,data[j,:])
#PLOT that compute the mean for each time point across all trials.
plt.plot(time,np.mean(data,axis=0),'k',linewidth=3)
plt.show()
```

```python
# get Fourier coefficients
dataX = scipy.fftpack.fft(data,axis=1) / pnts
hz = np.linspace(0,srate/2,int(np.floor(pnts/2)+1))#()

ave1 = 2 * abs(np.mean(dataX,axis=0))#Averaged coefficients
ave2=  np.mean( 2*np.abs(dataX),axis=0)#Averaged amplitude

```

```python
# plot both amplitude spectra
plt.stem(hz,ave1[0:len(hz)],'ks-',label='Average coefficients')
plt.stem(hz+.2,ave2[0:len(hz)],'ro-',label='Average amplitude')
plt.xlim([10,30])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Frequency domain')
plt.legend()
plt.show()
```
