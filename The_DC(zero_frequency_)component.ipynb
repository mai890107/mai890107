```python

import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Parameters
srate = 1000  # Sampling rate in Hz
time = np.arange(0., 2., 1/srate)  # Time vector (0 to 2 seconds)
pnts = len(time)  # Number of time points

# Create the signal
signal =1.5 + 2.5*np.sin( 2*np.pi*4*time )

fourTime = np.array(range(pnts))/pnts
fCoefs = np.zeros(len(signal),dtype = complex )

for fi in range(pnts):

  csw = np.exp( -1j*2*np.pi*fi*fourTime)
  fCoefs[fi] = np.sum(np.dot(signal,csw))

hz = np.linspace(0,srate/2,num=int(math.floor(pnts/2.)+1))
amplitude = 2*np.abs(fCoefs)
amplitude = amplitude/pnts
amplitude[0] = 1.5  # Directly set the DC component to the desired value


plt.stem(hz,amplitude[0:len(hz)])
plt.xlim(-.1,10)
plt.xlabel('Frequency (Hz)'), plt.ylabel('Amplitude (a.u.)')
plt.show()
```
